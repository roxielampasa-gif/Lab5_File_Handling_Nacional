1. Analyze and Compare File Content Complexity

```python
from pathlib import Path
import os

def analyze_complexity():
    # This points to C:\Users\YourName\Documents\Nacional_Activity_5
    file_path = Path.home() / "Documents" / "Nacional_Activity_5"
    
    # Check if the folder exists first to avoid the FileNotFoundError
    if not file_path.exists():
        print(f"Error: The folder {file_path} does not exist. Please create it!")
        return

    # Get all .txt files
    files = [f for f in os.listdir(file_path) if f.endswith('.txt')]
    
    if not files:
        print("Folder found, but no .txt files were detected.")
        return

    stats = []
    for file in files:
        # Use / to join paths with pathlib
        with open(file_path / file, 'r') as f:
            lines = [line for line in f.readlines() if line.strip()]
            content = " ".join(lines)
            words = content.split()
            
            num_lines = len(lines)
            num_words = len(words)
            num_chars = len(content)
            
            wpl = num_words / num_lines if num_lines > 0 else 0
            cpw = num_chars / num_words if num_words > 0 else 0
            
            stats.append([file, num_lines, num_words, num_chars, round(wpl, 2), round(cpw, 2)])

    stats.sort(key=lambda x: x[4], reverse=True)

    print(f"{'Filename':<25} | {'Lines':<5} | {'Words':<5} | {'Chars':<5} | {'W/L':<5} | {'C/W':<5}")
    print("-" * 75)
    for s in stats:
        print(f"{s[0]:<25} | {s[1]:<5} | {s[2]:<5} | {s[3]:<5} | {s[4]:<5} | {s[5]:<5}")

analyze_complexity()
```

2. Build a Context-Aware Word Frequency Analyzer

```python
from pathlib import Path
import string
from collections import Counter

def word_frequency_analyzer():
    # Set path using your logic
    file_path = Path.home() / "Documents" / "Nacional_Activity_5"
    
    # Common words to ignore (Stopwords)
    stopwords = {"the", "is", "and", "to", "in", "it", "of", "for", "with", "a", "an", "on", "this"}
    all_words = []

    if not file_path.exists():
        print(f"Directory not found: {file_path}")
        return

    # Process every .txt file in the folder
    files = list(file_path.glob("*.txt"))
    
    for file in files:
        with open(file, 'r') as f:
            # Normalize: lowercase and remove punctuation
            text = f.read().lower()
            text = text.translate(str.maketrans('', '', string.punctuation))
            
            # Filter words
            words = [w for w in text.split() if w not in stopwords]
            all_words.extend(words)

    # Count and display
    counts = Counter(all_words)
    print(f"\n--- Top 10 Meaningful Words in {file_path.name} ---")
    print(f"{'Word':<15} | {'Frequency':<10}")
    print("-" * 30)
    for word, count in counts.most_common(10):
        print(f"{word:<15} | {count:<10}")

word_frequency_analyzer()
```

3. Design a Smart File Backup System with History Tracking

```python
from pathlib import Path
import shutil
from datetime import datetime

def backup_system(target_file_name):
    # --- STUDENT DATA ---
    student_id = "TUPM-25-0363"
    student_name = "Nacional, Roxie"
    
    # Define paths using pathlib
    base_path = Path.home() / "Documents" / "Nacional_Activity_5"
    backup_dir = base_path / f"backup_{student_id}"
    source_file = base_path / target_file_name

    # 1. Create the folders if they don't exist
    base_path.mkdir(parents=True, exist_ok=True)
    backup_dir.mkdir(parents=True, exist_ok=True)

    # 2. Check if the source file exists
    if not source_file.exists():
        # Let's create a dummy file so you can actually see it work!
        with open(source_file, "w") as f:
            f.write("Sample content for Lab 7 backup exercise.")
        print(f"Note: Created '{target_file_name}' as it was missing.")

    # 3. Generate unique filename: {original}_{id}_{timestamp}.txt
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    new_filename = f"{source_file.stem}_{student_id}_{timestamp}{source_file.suffix}"
    dest_path = backup_dir / new_filename

    # 4. Perform the copy
    shutil.copy2(source_file, dest_path)
    file_size = dest_path.stat().st_size

    # 5. Update the personalized backup log
    log_file = backup_dir / f"backup_log_{student_id}.txt"
    with open(log_file, "a", encoding='utf-8') as log:
        log.write(f"Student: {student_name} ({student_id})\n")
        log.write(f"Original: {target_file_name} | Backup: {new_filename}\n")
        log.write(f"Timestamp: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}\n")
        log.write(f"Size: {file_size} bytes\n")
        log.write("-" * 50 + "\n")

    # Final Output
    print(f"Backup completed for {student_id} ({student_name})")
    print(f"File saved as: {new_filename}")
    print(f"Log updated at: {log_file}")

# --- THE CALL (This makes it run!) ---
backup_system("intro_2025XXXX.txt")
```
