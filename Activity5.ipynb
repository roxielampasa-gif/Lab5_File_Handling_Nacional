4. Safe File Operations with Backup

```python
# Step 1: Import additional libraries
from datetime import datetime
import shutil

# Step 2: Use your custom folder
backup_dir = Path.home() / "Documents" / "Nacional_Activity_5"
backup_dir.mkdir(exist_ok=True)

# Step 3: Function with "Try It" (Surname and Append/Overwrite choice)
def write_with_backup(filename: str, content: str, mode="w"):
    file_path = backup_dir / filename
    
    # Create backup if file exists
    if file_path.exists():
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        # Try It: Added Surname to the backup name
        backup_name = f"{file_path.stem}_Nacional_backup_{timestamp}{file_path.suffix}"
        backup_path = file_path.with_name(backup_name)
        shutil.copy2(file_path, backup_path)
        print(f"Backup saved: {backup_path.name}")
    
    # Try It: User choice for mode is handled by the 'mode' parameter
    with open(file_path, mode, encoding="utf-8") as file:
        file.write(content)
    print(f"File saved: {file_path.name} (Mode: {mode})")

# Try It: New function to list backups for a specific file
def list_specific_backups(filename: str):
    print(f"\n--- Listing backups for: {filename} ---")
    stem = Path(filename).stem
    # Searches for files that start with the filename and contain 'backup'
    for backup in backup_dir.glob(f"{stem}*backup*"):
        print("-", backup.name)

# --- Run Demonstration ---
print("=== File Operations Demo ===")

# 1. Create original
write_with_backup("demo.txt", "This is the first version.")

# 2. Update (This triggers the backup)
write_with_backup("demo.txt", "\nThis is the second version.", mode="a") # 'a' for append

# 3. List only backups for demo.txt
list_specific_backups("demo.txt")
```

5. Menu-Driven File Manager

```python
# Step 1: Import required modules
from pathlib import Path
from datetime import datetime
import shutil

# Try It: Change working directory to Surname_Activity_5
backup_dir = Path.home() / "Documents" / "Nacional_Activity_5"
backup_dir.mkdir(exist_ok=True)

def file_manager():
    file_name = input("Enter filename (e.g., notes.txt, data.csv, or readme.md): ")
    file_path = backup_dir / file_name
    
    while True:
        print("\n--- MENU ---")
        print("1. Write to file")
        print("2. Append to file")
        print("3. Read file")
        print("4. Backup file (with Surname)")
        print("5. List all backup files") # Added 6th option logic
        print("6. Exit")
        
        choice = input("Choose an option (1-6): ")
        
        if choice == "1":
            content = input("Enter content to write:\n")
            with open(file_path, "w", encoding="utf-8") as f:
                f.write(content)
            print("File written successfully.")
            
        elif choice == "2":
            more = input("Enter content to append:\n")
            with open(file_path, "a", encoding="utf-8") as f:
                f.write("\n" + more)
            print("Content appended.")
            
        elif choice == "3":
            if file_path.exists():
                with open(file_path, "r", encoding="utf-8") as f:
                    print(f"\n--- Content of {file_name} ---\n", f.read())
            else:
                print("File not found.")
                
        elif choice == "4":
            if file_path.exists():
                timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
                # Try It: Modify backup name to include Surname
                backup_file = file_path.with_name(
                    f"{file_path.stem}_Nacional_backup_{timestamp}{file_path.suffix}"
                )
                shutil.copy2(file_path, backup_file)
                print(f"Backup created: {backup_file.name}")
            else:
                print("Cannot backup. File does not exist.")
        
        elif choice == "5":
            # Try It: 6th menu option to list all backup files
            print("\n--- List of All Backups ---")
            backups = list(backup_dir.glob("*backup*"))
            if backups:
                for b in backups:
                    print(f"- {b.name}")
            else:
                print("No backups found.")
                
        elif choice == "6":
            print("Exiting the file manager.")
            break
        else:
            print("Invalid choice. Please try again.")

# Run the tool
if __name__ == "__main__":
    file_manager()
```

2. Creating a Timestamped Backup Automatically

```python
from datetime import datetime
import shutil
from pathlib import Path

def create_backup(original_path):
    if Path(original_path).exists():
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        backup_name = f"{original_path.stem}_backup_{timestamp}{original_path.suffix}"
        backup_path = original_path.parent / backup_name
        shutil.copy(original_path, backup_path)
        print(f"Backup created: {backup_path}")
    else:
        print("No original file found to back up.")
```

```python
from pathlib import Path

student_id = "TUPM-25-0363"
# Ensure this folder exists and has your .txt files inside!
target_dir = Path.home() / "Documents" / "Nacional_Activity_5"

files = list(target_dir.glob("*.txt"))
stats = []

print(f"{'Filename':<25} | {'Lines':<5} | {'Words':<5} | {'Chars':<5} | {'W/L':<5} | {'C/W':<5}")
print("-" * 70)

for f_path in files:
    content = f_path.read_text(encoding="utf-8")
    lines = [line for line in content.splitlines() if line.strip()]
    words = content.split()
    chars = len(content.replace(" ", "").replace("\n", ""))
    
    num_lines = len(lines) if len(lines) > 0 else 1
    num_words = len(words)
    
    wpl = num_words / num_lines
    cpw = chars / num_words if num_words > 0 else 0
    
    stats.append({
        'name': f_path.name,
        'lines': num_lines,
        'words': num_words,
        'chars': chars,
        'wpl': wpl,
        'cpw': cpw
    })
    
    print(f"{f_path.name:<25} | {num_lines:<5} | {num_words:<5} | {chars:<5} | {wpl:<5.1f} | {cpw:<5.1f}")

# Sort by word density (WPL)
stats.sort(key=lambda x: x['wpl'], reverse=True)
if stats:
    print(f"\nMost dense file: {stats[0]['name']} ({stats[0]['wpl']:.1f} words/line)")
```

```python
from pathlib import Path
from datetime import datetime
import shutil

# --- Configuration ---
student_id = "TUPM-25-0363"
student_name = "Nacional"
base_dir = Path.home() / "Documents" / "Nacional_Activity_5"
backup_dir = base_dir / f"backup_{student_id}"
backup_dir.mkdir(exist_ok=True)

def smart_backup(filename):
    original_file = base_dir / filename
    if not original_file.exists():
        print(f"Error: {filename} not found.")
        return

    # 1. Generate Timestamp and Paths
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    backup_name = f"{original_file.stem}_{student_id}_{timestamp}{original_file.suffix}"
    dest_path = backup_dir / backup_name

    # 2. Copy File
    shutil.copy2(original_file, dest_path)
    file_size = dest_path.stat().st_size

    # 3. Update Log File
    log_file = backup_dir / f"backup_log_{student_id}.txt"
    with open(log_file, "a", encoding="utf-8") as log:
        log.write(f"--- Backup Entry ---\n")
        log.write(f"Student: {student_name} ({student_id})\n")
        log.write(f"File: {filename} | Timestamp: {timestamp}\n")
        log.write(f"Size: {file_size} bytes | Path: {dest_path}\n\n")

    print(f"Backup completed for {student_id} ({student_name})")
    print(f"File saved as: {backup_name}")
    print(f"Log updated at: {log_file.name}")

# Test it:S
# smart_backup("intro_TUPM-25-0363.txt")
```
